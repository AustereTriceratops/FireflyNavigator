```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
data = pd.read_csv("xyzt_tent.csv", names=["x", "y", "z", "time"])  # float64, float64, float64, int64
data.head(5)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8,10))

data.plot("x", "y", kind="scatter", ax=ax[0], c="z")
data.plot("z", "x", kind="scatter", ax=ax[1], c="y")
data.plot("y", "z", kind="scatter", ax=ax[2], c="x")
```

```python
# find last timepoint

t = data['time'][len(data.index)-1]
t
```

```python
#find number of flashing events per timestep

activity_counts = data["time"].value_counts().sort_index()
activity_counts
```

```python
# pad for timesteps with no activity

padded_activity_counts = [0 for _ in range(t+1)]

for c in activity_counts.keys():
    padded_activity_counts[c] = activity_counts[c]
    
padded_activity_counts = pd.Series(padded_activity_counts)
```

```python
# histogram

ax = sns.distplot(padded_activity_counts, kde=False, norm_hist=True)
ax.set(xlabel="number of events in a given timestep", ylabel="frequency")

plt.show()
```

```python
# find how active fireflies are over time

fig, ax = plt.subplots(figsize=(16, 12))

ax = sns.lineplot(padded_activity_counts.index[:10000], padded_activity_counts[:10000])
ax.set(xlabel="timestep", ylabel="activity density")
plt.show()
```

```python
# what does the distribution of gaps between the bursts of activity look like?

gaps = []
s = 0
active = True

for val in padded_activity_counts:
    if val == 0:
        s += 1
        active = False
    else:
        if not active:
            if s > 100:  # arbitrary cutoff to find large gaps
                gaps.append(s)
            s = 0
            active = True
            
sns.distplot(gaps, bins=16, norm_hist=True, kde=False)
```

```python
# there are smaller gaps in activity in the middle of bursts
# it seems like the system operates on two different timescales (?)

gaps = []
s = 0
active = True

for val in padded_activity_counts:
    if val == 0:
        s += 1
        active = False
    else:
        if not active:
            if s <= 100:  # arbitrary cutoff to find small gaps
                gaps.append(s)
            s = 0
            active = True
            
sns.distplot(gaps, bins=16, norm_hist=True, kde=False)
```

```python
# tracking a single firefly

def find_velocity(df, df_index):
    p_0 = np.array(data.loc[df_index])
    
    # look at next 20 entries in dataframe
    # choose the point which is closest to p_0
    # this is assumed to be where the firefly flew to next
    # this fails when two or more fireflies are close together
    
    min_dist = np.Inf
    p_f = None
    
    for i in range(20):
        p_i = np.array(data.loc[df_index + i + 1])
                   
        distance = np.sqrt(np.sum(np.square(p_i[:3] - p_0[:3])))
        
        if distance < min_dist:
            min_dist = distance
            p_f = p_i
    
    return min_dist/(p_f[3] - p_0[3])
```

```python
x = [i for i in range(10)]
y = [find_velocity(data, i) for i in x]

plt.plot(x, y)
```
